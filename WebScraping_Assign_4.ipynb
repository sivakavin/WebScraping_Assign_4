### 1. Scrape the details of most viewed videos on YouTube from Wikipedia:
    Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos/
    
    You need to find following details:
        a) Rank
        B) Name
        C) Artist
        D) Upload date
        E) Views

```python
# Access chrome browser using webdrive
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(3)


#Access the website
url = "https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos"
driver.get(url)
time.sleep(3)


# creating empty dictionary
most_Viewed_youtube = {}
most_Viewed_youtube['Rank'] = []
most_Viewed_youtube['Name'] = []
most_Viewed_youtube['Artist'] = []
most_Viewed_youtube['Upload_date'] = []
most_Viewed_youtube['Views'] = []
time.sleep(3)


a = 1 #counter value


# Scraping the all data
for i in range(0,30):
    time.sleep(3)
    
    try:
        #rank
        rank = driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[1]".format(a))
        most_Viewed_youtube['Rank'].append(rank.text)
        time.sleep(2)
            
        #name
        name = driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[2]".format(a))
        most_Viewed_youtube['Name'].append(name.text)
        time.sleep(2)
        
        #artist
        artist = driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[3]".format(a))
        most_Viewed_youtube['Artist'].append(artist.text)
        time.sleep(2)
    
        #upload_date
        uploadDate = driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[5]".format(a))
        most_Viewed_youtube['Upload_date'].append(uploadDate.text)
        time.sleep(2)
    
        #views
        view = driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr[{}]/td[4]".format(a))
        most_Viewed_youtube['Views'].append(view.text)
        time.sleep(2)
    except NoSuchElementException as n:
        print("Exception",n)
   
    # incresing counter variable
    a += 1


Most_Viewed_youtube = pd.DataFrame(most_Viewed_youtube)
Most_Viewed_youtube
```

*Note: - From bcci.tv home page you have reach to the international fixture page through code.*

```python
# program for list to string converstion

def list_to_string(value):
    
    stri = '' #empty string intillation
    #return value
    for i in value:
        stri += i  #appending the value
        
    return stri    
```

### 5. Scrape the details of trending repositories on Github.com.
       Url = https://github.com/
       You have to find the following details:
    A) Repository title
    B) Repository description
    C) Contributors count
    D) Language used
 
Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
# library
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException

# Accessing chrome browser
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(4)

# navigating to webpage
url ="https://github.com/"
driver.get(url)
time.sleep(7)

# navigating to webpage
explore_opt = driver.find_element_by_xpath("//ul[@class='d-lg-flex list-style-none']/li[4]")
action = ActionChains(driver)
action.click_and_hold(explore_opt).perform()

#clicking trending option
trend_opt = driver.find_element_by_xpath("//ul[@class='list-style-none mb-3'][2]/li[3]/a")
action.click(trend_opt).perform()

# Creating empty dictionary to append the data
trending_repo = {}
trending_repo['Repository_Title'] = []
trending_repo['Repository_Description'] = []
trending_repo['Contributors_Count'] = []
trending_repo['Language_Used'] = []

# repo_title
titles = driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
for i in titles:
    if i.text == " ":
        trending_repo['Repository_Title'].append("-")
    else:
        trending_repo['Repository_Title'].append(i.text)
time.sleep(2)

#description
count = 1
for i in range(len(driver.find_elements_by_xpath("//article[@class='Box-row']"))):
    
    try:
        descriptions = driver.find_element_by_xpath(f"//article[@class='Box-row'][{count}]/p[1]")
        trending_repo['Repository_Description'].append(descriptions.text) 
        
    except NoSuchElementException:
        trending_repo['Repository_Description'].append("-")
    
    count += 1

#contributor_counts
contributor_counts = driver.find_elements_by_xpath("//div[@class='f6 color-text-secondary mt-2']/a[2]")
for i in contributor_counts:
    if i.text == "":
        trending_repo['Contributors_Count'].append("-")
    else:
        trending_repo['Contributors_Count'].append(i.text)
time.sleep(2)

#language
lan_count = 1
for i in range(len(driver.find_elements_by_xpath("//div[@class='f6 color-text-secondary mt-2']"))):
    try:
        langu = driver.find_element_by_xpath(f"//article[@class='Box-row'][{lan_count}]/div[2]/span/span[2]")
        trending_repo['Language_Used'].append(langu.text) 
        
    except NoSuchElementException:
        trending_repo['Language_Used'].append("-")
    lan_count += 1
time.sleep(2)


#creating dataframe
Trending_repositories = pd.DataFrame(trending_repo)
Trending_repositories.to_csv("Trending repositories.csv")
Trending_repositories.head()
```

### 6. Scrape the details of top 100 songs on billboard.com.

            B) Artist name
            C) Last week rank
            D) Peak rank
            E) Weeks on board
        
**Note: - From the home page you have to click on the charts option then hot 100-page link through code. **

```python
# library
from selenium.webdriver.common.action_chains import ActionChains
from selenium.common.exceptions import ElementNotInteractableException,ElementNotSelectableException,ElementClickInterceptedException


# Accessing chrome browser
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(4)

# navigating to webpage
url ="https://www.billboard.com/"
driver.get(url)
time.sleep(7)

# Navigating to hot100 page
chart_opt=driver.find_element_by_xpath("//*[@id='root']/div[2]/div[1]/div[1]/ul/li[1]/a").get_attribute('href')
driver.get(chart_opt)
time.sleep(5)

hot_opt = driver.find_element_by_xpath("//*[@id='main']/div[2]/div/div[1]/a").get_attribute('href')
driver.get(hot_opt)
time.sleep(5)


#Creating empty list
top100song = {}
top100song["SongName"] = []
top100song["ArtistName"] = []
top100song["Last_week_rank"] = []
top100song["Peak_rank"] = []
top100song["Weeks_on_board"] = []

#Scraping required detail
songName = driver.find_elements_by_xpath("//span[@class='chart-element__information']/span[1]")
for i in songName:
    top100song["SongName"].append(i.text)

artistName = driver.find_elements_by_xpath("//span[@class='chart-element__information']/span[2]")
for i in artistName:
    top100song["ArtistName"].append(i.text)
    
lastWeek = driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--last']")
for i in lastWeek:
    top100song["Last_week_rank"].append(i.text)
    
peakRank = driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--peak']")
for i in peakRank:
    top100song["Peak_rank"].append(i.text)
 
weekONBoard = driver.find_elements_by_xpath("//div[@class='chart-element__meta text--center color--secondary text--week']")
for i in weekONBoard:
    top100song["Weeks_on_board"].append(i.text)
    
# Dataframe
Top100Song = pd.DataFrame(top100song)
Top100Song.to_csv("Top100Song.csv")
Top100Song.head(15)

#after all process closing the chrome browser
driver.close()
```

    **You have to find the following details:**
    
    A) Book name
    B) Author name
    C) Volumes sold
    D) Publisher
    E) Genre

```python
# Accessing chrome browser
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(3)

url = "https://www.theguardian.com/news/datablog/2012/aug/09/best-selling-books-all-time-fifty-shades-grey-compare/"
driver.get(url)
time.sleep(3)

# creation dictionart files to store the scraped detail
Novels = {}
Novels["Book_Name"] = []
Novels["Author_Name"] = []
Novels["Volume_Sold"] = []
Novels["Publisher"] = []
Novels["Genre"] = []

#Scraping the detail

count = 1
for i in range(len(driver.find_elements_by_xpath("//table[@class='in-article sortable']/tbody/tr"))):
    
    book_name = driver.find_element_by_xpath(f"//table[@class='in-article sortable']/tbody/tr[{count}]/td[2]")
    Novels["Book_Name"].append(book_name.text)
    
    author_name = driver.find_element_by_xpath(f"//table[@class='in-article sortable']/tbody/tr[{count}]/td[3]")
    Novels["Author_Name"].append(author_name.text)
    
    volume_name = driver.find_element_by_xpath(f"//table[@class='in-article sortable']/tbody/tr[{count}]/td[4]")
    Novels["Volume_Sold"].append(volume_name.text)
    
    publisher = driver.find_element_by_xpath(f"//table[@class='in-article sortable']/tbody/tr[{count}]/td[5]")
    Novels["Publisher"].append(publisher.text)
    
    genre = driver.find_element_by_xpath(f"//table[@class='in-article sortable']/tbody/tr[{count}]/td[6]")
    Novels["Genre"].append(genre.text)
    
    count += 1
    
Novels_Topselling = pd.DataFrame(Novels)
Novels_Topselling.to_csv("Novels_Topselling.csv")
Novels_Topselling.head(15)
```

### 9. Scrape the details most watched tv series of all time from imdb.com.
    Url = https://www.imdb.com/list/ls095964455/
**You have to find the following details:**

    A) Name
    B) Year span
    C) Genre
    D) Run time
    E) Ratings
    F) Votes

```python
# Accessing chrome browser
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(3)

url = "https://www.imdb.com/list/ls095964455/"
driver.get(url)
time.sleep(3)

# creating empty disctionary
imdb_series = {}
imdb_series['Name'] = []
imdb_series['YearSpan'] = []
imdb_series['Genre'] = []
imdb_series['RunTime'] = []
imdb_series['Rating'] = []
imdb_series['Votes'] = []

#Scraping the above data from IMDB Website
names = driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a")
for i in names:
    if i.text is None:
        imdb_series['Name'].append("-")
    else:
        imdb_series['Name'].append(i.text)
        
yearSpan = driver.find_elements_by_xpath("//h3[@class='lister-item-header']/span[2]")
for i in yearSpan:
    if i.text is None:
        imdb_series['YearSpan'].append("-")
    else:
        imdb_series['YearSpan'].append(i.text.replace('(','').replace(')',''))

        
genre = driver.find_elements_by_xpath("//p[@class='text-muted text-small']/span[5]")
for i in genre:
    if i.text is None:
        imdb_series['Genre'].append("-")
    else:
        imdb_series['Genre'].append(i.text)
        
runtime = driver.find_elements_by_xpath("//p[@class='text-muted text-small']/span[3]")
for i in runtime:
    if i.text is None:
        imdb_series['RunTime'].append("-")
    else:
        imdb_series['RunTime'].append(i.text)
        
        
rating = driver.find_elements_by_xpath("//div[@class ='ipl-rating-star small']/span[2]")
for i in rating:
    if i.text is None:
        imdb_series['Rating'].append("-")
    else:
        imdb_series['Rating'].append(i.text)

votes = driver.find_elements_by_xpath("//div[@class='lister-item-content']/p[4]/span[2]")
for i in votes:
    if i.text is None:
        imdb_series['Votes'].append("-")
    else:
        imdb_series['Votes'].append(i.text)        

    
# Creating dataframe using pandas lib
IMDB_Series = pd.DataFrame(imdb_series)
IMDB_Series.to_csv("IMDB_Series.csv")

IMDB_Series.head(15)
```

### 10. Details of Datasets from UCI machine learning repositories.
    Url = https://archive.ics.uci.edu/

**You have to find the following details:**

    B) Data type
    C) Task
    D) Attribute type
    E) No of instances
    F) No of attribute
    G) Year

#### Note: - from the home page you have to go to the Show All Dataset page through code.

```python
# Accessing chrome browser
driver = webdriver.Chrome("chromedriver.exe")
time.sleep(3)

url = "https://archive.ics.uci.edu/"
driver.get(url)
time.sleep(3)

# navigating to view all dataset page
view_dataset = driver.find_element_by_xpath("/html/body/table[1]/tbody/tr/td[2]/span[2]/a/font/b")
view_dataset.click()
time.sleep(3)

# creating empty disctionart and nested list to append scra[p]
uci_repositories = {}
uci_repositories['Dataset name']= []
uci_repositories['Data type'] = []
uci_repositories['Task'] = []
uci_repositories['Attribute type'] = []
uci_repositories['No of instances'] = []
uci_repositories['No of attribute'] = []
uci_repositories['Year'] = []


# Scraping the required detail

name = driver.find_elements_by_xpath("//p[@class='normal']/b/a")
for i in name:
    if len(uci_repositories['Dataset name']) < 586:
        uci_repositories['Dataset name'].append(i.text)
    else:
        pass
     
time.sleep(1)
for i in range(2,588):
    dtype = driver.find_element_by_xpath(f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[2]/p")
    if dtype.text == " ":
        uci_repositories['Data type'].append("-")
    else:
        uci_repositories['Data type'].append(dtype.text)
    
    task = driver.find_element_by_xpath(f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[3]/p")
    if task.text == " ":
        uci_repositories['Task'].append("-")
    else:
        uci_repositories['Task'].append(task.text) 
        
    type = driver.find_element_by_xpath(f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[4]/p")
    if type.text == " ":
        uci_repositories['Attribute type'].append("-")
    else:
        uci_repositories['Attribute type'].append(type.text)
        
    instance = driver.find_element_by_xpath(f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[5]/p")
    if instance.text == " ":
        uci_repositories['No of instances'].append("-")
    else:
        uci_repositories['No of instances'].append(instance.text) 
        
    num_attribute = driver.find_element_by_xpath(f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[6]/p")
    if num_attribute.text == " ":
        uci_repositories['No of attribute'].append("-")
    else:
        uci_repositories['No of attribute'].append(num_attribute.text)        
        
    year = driver.find_element_by_xpath(f"/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr[{i}]/td[7]/p")
    if year.text == " ":
        uci_repositories['Year'].append("-")
    else:
        uci_repositories['Year'].append(year.text)
time.sleep(1)


#Creating Dataframe
UCI_Repositories = pd.DataFrame(uci_repositories)
UCI_Repositories.to_csv("UCI_Repositories.csv")
UCI_Repositories.head(20)
```
